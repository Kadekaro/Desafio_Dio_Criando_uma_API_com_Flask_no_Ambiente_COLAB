# **No código abaixo, eu consigo resolver o desafio que me foi passado de forma simples, transformando um arquivo do tipo dicionário em DataFrame e depois em formato HTML:**

```python
from flask import Flask
from flask_ngrok import run_with_ngrok
import pandas as pd

app = Flask(__name__)
run_with_ngrok(app)

table = {
    "Number": [1, 2, 3, 4, 5],
    "Name": ["Mahesh", "Alex",  "David", "John", "Chris"],
    "Age": [25, 26, 27, 28, 29],
    "City": ["Bangalore", "London", "San Francisco", "Toronto", "Paris"],
    "Country": ["India", "UK", "USA", "Canada", "France"]
}

df = pd.DataFrame(table)

@app.route("/")
def home():
    return "<marquee><h3> Bem-vindo à Página Inicial! </h3></marquee>"

@app.route("/input")
def input_data():
    return df.to_html()

@app.route('/output', methods=['GET', 'POST'])
def output():
    return df.to_html()

if __name__ == "__main__":
    app.run()
```

Em baixo, eu utilizo um template html para criar a tabela. É uma forma eficaz, porém é mais difícil já que exige domínio de outras linguagens, fiz mais pelo desafio e pela aprendizagem, mas prefiro o método acima com o pandas ^^.

```python
from flask import Flask, jsonify, render_template
from flask_ngrok import run_with_ngrok

modelo = '/content/drive/MyDrive/Colab Notebooks'
app = Flask(__name__, template_folder=modelo)
run_with_ngrok(app)

table = {
    "Number": [1, 2, 3, 4, 5],
    "Name": ["Mahesh", "Alex", "David", "John", "Chris"],
    "Age": [25, 26, 27, 28, 29],
    "City": ["Bangalore", "London", "San Francisco", "Toronto", "Paris"],
    "Country": ["India", "UK", "USA", "Canada", "France"]
}

@app.route("/")
def home():
    return "<marquee><h3> Bem-vindo à Página Inicial! </h3></marquee>"

@app.route("/input")
def input_data():
    return jsonify(table)

@app.route('/output')
def output():
    return render_template('representando_o_json_em_tabela.html', data=table)

if __name__ == "__main__":
    app.run()
```
